```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_x = pd.read_csv("/content/Iris Dataset.csv")
df_x
```

```python
df_y = pd.DataFrame(df_x['Species'])
df_y
```

```python
df_x = df_x.drop('Species', axis = 1)
df_x = df_x.drop('Id', axis = 1)
df_x
```

```python
# Label Encoding

species = {"Iris-setosa": 0, "Iris-versicolor": 1, "Iris-virginica": 2}
df_y['Species'] = df_y['Species'].map(species)

df_y
```

```python
# Normalisation

df_x_norm = (df_x - df_x.min()) / (df_x.max() - df_x.min())
df_y_norm = (df_y - df_y.min()) / (df_y.max() - df_y.min())

df_x_norm = df_x_norm.values
df_y_norm = df_y_norm.values

df_x_norm
```

```python
df_y_norm
```

```python
def k_means_clustering(dataset, species, num_clusters = 3):
    # Randomly initializing cluster centroids
    centroids_indices = np.random.choice(dataset.shape[0], num_clusters, replace=False)
    centroids = dataset[centroids_indices]
    
    # Initializing variables
    max_iterations = 500
    num_examples = dataset.shape[0]
    num_features = dataset.shape[1]
    cluster_assignments = np.zeros(num_examples)
    
    for k in range(max_iterations):
        # Assigning each example to the nearest centroid
        for i in range(num_examples):
            distances = np.linalg.norm(dataset[i] - centroids, axis=1)
            cluster_assignments[i] = np.argmin(distances)
        
        # Updating centroids
        for j in range(num_clusters):
            cluster_examples = dataset[cluster_assignments == j]
            centroids[j] = np.mean(cluster_examples, axis=0)
    
    # Plotting the output of the cluster and actual species
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
    ax1.scatter(dataset[:, 0], dataset[:, 1], c = cluster_assignments)
    ax1.set_title('K-Means Clustering Output')
    ax2.scatter(dataset[:, 0], dataset[:, 1], c = species)
    ax2.set_title('Actual Species')
    plt.show()
```

```python
k_means_clustering(df_x_norm, df_y_norm)
```

```python
def principal_component_analysis(X, species):
    
    # Calculating the covariance matrix
    covariance_matrix = np.cov(X.T)
    
    # Calculating the eigenvalues and eigenvectors
    eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)
    
    # Sorting the eigenvectors by eigenvalues in descending order
    eigenvectors = eigenvectors[:, np.argsort(eigenvalues)[::-1]]
    eigenvalues = eigenvalues[np.argsort(eigenvalues)[::-1]]
    
    # Projecting the data onto the first three eigenvectors
    X_pca = np.dot(X, eigenvectors[:, :3])
    
    # Plotting the data in the first three eigenvectors
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c = species)
    ax.set_title('Principal Component Analysis')
    plt.show()
    
    return eigenvalues[:3], eigenvectors[:, :3]
```

```python
eigenvalues, eigenvectors = principal_component_analysis(df_x_norm, df_y_norm)
print('PCA Eigenvalues:', eigenvalues)
print('PCA Eigenvectors:\n', eigenvectors)
```
